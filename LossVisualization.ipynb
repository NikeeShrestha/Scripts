```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
training=pd.read_csv('Losses/Training_loss.csv', index_col=0)
val = pd.read_csv('Losses/Validation_loss.csv', index_col=0)
```

```python
merged = pd.merge(training, val, left_index=True, right_index=True)
```

```python
# merged
```

```python
plt.plot('total_loss_x', data=merged, label = 'Total Training Loss')
plt.plot('total_loss_y', data=merged, label = 'Total Validation Loss')
plt.legend()
plt.xlabel('epochs')
plt.savefig('Total_loss.png')
```

```python
plt.plot('barlow_loss_x', data=merged, label ='Barlow Training Loss')
plt.plot('barlow_loss_y', data=merged, label = 'Barlow Validation Loss')
plt.legend()
plt.xlabel('epochs')
plt.savefig('barlow_loss.png')
```

```python
plt.plot('mse_loss_x', data=merged, label = 'MAE Training Loss')
plt.plot('mse_loss_y', data=merged, label = 'MAE Validation Loss')
plt.legend()
plt.xlabel('epochs')
plt.savefig('MAE_loss.png')
```
