```python
category=pd.read_csv("~/Documents/GenePrediction/All_data/Categorical.csv", index_col=0)
```

```python
category
```

```python
category_encoded=pd.DataFrame()
for i in category.columns:
    onehotencoder=OneHotEncoder(sparse_output=False)
    encodedfeatures=onehotencoder.fit_transform(category[[str(i)]])
    encoded_feature_names = onehotencoder.get_feature_names_out([str(i)])
    
    category_encoded = category_encoded.join(pd.DataFrame(encodedfeatures, columns=encoded_feature_names), how='outer')
```

```python
category_encoded=category_encoded.set_index(fulldf.index)
```

```python
fulldf=pd.merge(fulldf, category_encoded, left_index=True, right_index=True)
```

```python
fulldf=fulldf.drop('classical_label', axis=1)
```
