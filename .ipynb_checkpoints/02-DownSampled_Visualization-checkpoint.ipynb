```python
df=pd.read_csv("../../GenePrediction_files/Downsampled_data/Merged_Downsampled.csv", index_col=0)
```

```python
df=df.drop(["final_label","core_label"], axis=1)
df=df.dropna()
```

```python
df.visible_label.value_counts()
```

```python
scaler = preprocessing.StandardScaler()
scaled_data = scaler.fit_transform(df.drop(["visible_label","effect"],axis=1))

pca=PCA(n_components=2)
embeddings=pca.fit_transform(scaled_data)
np.cumsum(pca.explained_variance_)
```

```python
embeddings=pd.DataFrame(data = embeddings,columns = ['PC1', 'PC2'])
embeddings.index=df.index
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
```

```python
final_df=pd.concat([df,pd.DataFrame(data = embeddings, columns = ['PC1', 'PC2'])], axis=1)
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.pipeline import make_pipeline
import umap.plot 

# Sample the data - 100k
# tps_sample = tps.sample(100000)
X, y = df.drop(["visible_label", "effect"], axis=1), df[["visible_label"]].values.flatten()
```

```python
# scaler = preprocessing.StandardScaler()
# scaled_data = scaler.fit_transform(X)
```

```python
pipe = make_pipeline(PowerTransformer())
X = pipe.fit_transform(X.copy())
```

```python
manifold = umap.UMAP(n_neighbors=70).fit(X)
# X_reduced_2=manifold.transform(X)
```

```python
 # pip install umap-learn[plot]

umap.plot.points(manifold, labels=df.visible_label,theme="fire")

# plt.savefig('../Umap_syntenic_label.png')
```

```python
X, y = df.drop(["final_label", "core_label",'core_label_Dispensable Gene', 'core_label_Near-Core Gene',
       'core_label_Private Gene'], axis=1), df[["core_label"]].values.flatten()
pipe = make_pipeline(PowerTransformer())
X = pipe.fit_transform(X.copy())
```

```python
y_encoded = pd.factorize(y)[0]
```

```python
%%time

manifold = umap.UMAP(n_neighbors=70).fit(X,y_encoded)
```

```python
umap.plot.points(manifold, labels=df.Sytenic, theme="fire");
plt.savefig('../Umap_core_fitted_syntenic_label.png')
```

```python
umap.plot.points(manifold, labels=y, theme="fire");
plt.savefig('../Umap_core_fitted_core_label.png')
```

```python
X, y = df.drop(["final_label",'core_label',"Sytenic"], axis=1), df[["Sytenic"]].values.flatten()
pipe = make_pipeline(PowerTransformer())
X = pipe.fit_transform(X.copy())
```

```python
%%time

# manifold = umap.UMAP(n_neighbors=70).fit(X,y)
umap.plot.points(manifold, labels=df.final_label, theme="fire");


plt.savefig('../Umap_syntenic_fitted_final_label.png')
```

```python
umap1=pd.DataFrame(data = manifold.embedding_,columns = ['umap1', 'umap2'])
umap1.index=df.index
final_df=pd.concat([df,pd.DataFrame(data = umap1, columns = ['umap1', 'umap2'])], axis=1)
```

```python
fig=sns.kdeplot(x=final_df.loc[final_df['final_label']==1,:]['umap1'],y=final_df.loc[final_df['final_label']==1,:]['umap2'],fill=True,color="darkblue", alpha=0.8,label='Validated Maize Gene Model')
# plt.contour(final_df.loc[final_df['final_label']==1,:]['PC1'],final_df.loc[final_df['final_label']==1,:]['PC2'], colors='darkblue', linewidths=10)
fig=sns.kdeplot(x=final_df.loc[final_df['final_label']==0,:]['umap1'],y=final_df.loc[final_df['final_label']==0,:]['umap2'], fill=True,color="maroon", alpha=0.8,label='Other Maize Gene Models')
plt.xlim([0,25])
plt.ylim([1,12.5])


custom_legend_handles = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Validated Maize Gene Models',alpha=0.5),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Other Maize Gene Models',alpha=0.5)
]

plt.legend(handles=custom_legend_handles, loc='lower left')
plt.savefig('../umapkdeplot.png', bbox_inches='tight')
plt.show()
```
