```python
import pandas as pd
# pd.set_option('mode.use_inf_as_null', False)
import numpy as np
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
import xgboost
from numpy import mean
from numpy import std
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr
from math import sqrt
import glob
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from scipy.stats import randint
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from matplotlib.lines import Line2D
```

```python
df=pd.read_csv("../../GenePrediction_files/Downsampled_data/Merged_Downsampled.csv", index_col=0)
```

```python
df=df.dropna()
```

```python
df.final_label.value_counts()
```

```python
scaler = preprocessing.StandardScaler()
scaled_data = scaler.fit_transform(df.drop(["final_label","core_label"],axis=1))

pca=PCA(n_components=2)
embeddings=pca.fit_transform(scaled_data)
np.cumsum(pca.explained_variance_)
```

```python
embeddings=pd.DataFrame(data = embeddings,columns = ['PC1', 'PC2'])
embeddings.index=df.index
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
```

```python
final_df=pd.concat([df,pd.DataFrame(data = embeddings, columns = ['PC1', 'PC2'])], axis=1)
```

```python
features = ['avg_gene_abundance', 'gene_breadth', 'exonLength', 'genelength','Exon_quad']
# final_df[['PC1','PC2']]=final_df[['PC1','PC2']]+1

fig=sns.kdeplot(x=final_df.loc[final_df['final_label']==1,:]['PC1'],y=final_df.loc[final_df['final_label']==1,:]['PC2'],fill=True,color="darkblue", alpha=0.8,label='Validated Maize Gene Model')
# plt.contour(final_df.loc[final_df['final_label']==1,:]['PC1'],final_df.loc[final_df['final_label']==1,:]['PC2'], colors='darkblue', linewidths=10)
fig=sns.kdeplot(x=final_df.loc[final_df['final_label']==0,:]['PC1'],y=final_df.loc[final_df['final_label']==0,:]['PC2'], fill=True,color="maroon", alpha=0.8,label='Other Maize Gene Models')
plt.xlim([-10,10])
plt.ylim([-10,14])


custom_legend_handles = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Validated Maize Gene Models',alpha=0.5),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Other Maize Gene Models',alpha=0.5)
]

plt.legend(handles=custom_legend_handles)

plt.savefig("../PC1_PC2_downsampled.png", bbox_inches='tight')
plt.show()
```
